# Cleaning

```python
df['Transaction ID'].duplicated().sum()
```

```python
 # Convert time columns to datetime
time_cols = ["Date of Journey","Date of Purchase", "Departure Time", "Arrival Time", "Actual Arrival Time","Time of Purchase"]
for col in time_cols:
    if col in df.columns:
          df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=False)
```

```python
df.dtypes
```

```python
df.columns.tolist()
```

```python
# Create Delay Duration (minutes)
if "Arrival Time" in df.columns and "Actual Arrival Time" in df.columns:
        df["Delay Duration (min)"] = (df["Actual Arrival Time"] - df["Arrival Time"]).dt.total_seconds() / 60
        df["Delay Duration (min)"] = df["Delay Duration (min)"].fillna(0)
```

```python
df.loc[df["Delay Duration (min)"] < 0, "Delay Duration (min)"] = 0
```

```python
#  Extract Day of Week & Month
if "Date of Journey" in df.columns:
    df["Day of Week"] = df["Date of Journey"].dt.day_name()
    df["Month"] = df["Date of Journey"].dt.month_name()
```

```python
if "Date of Journey" in df.columns:
    df["Year"] = df["Date of Journey"].dt.year
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df['Day of Week'] = pd.Categorical(df['Day of Week'], categories=day_order, ordered=True)
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
```

```python
df['Transaction ID'] = df['Transaction ID'].astype(str)
```

```python
df.dtypes
```

```python
df['Delay Duration (min)'] = df['Delay Duration (min)'].fillna(0).astype(int)

df['Year'] = df['Year'].astype('Int64')
```

```python
mask = df['Actual Arrival Time'] != 'Cancelled'
df.loc[mask, 'Actual Arrival Time'] = pd.to_datetime(df.loc[mask, 'Actual Arrival Time'], errors='coerce')
```

```python
df.dtypes
```

```python
# Converting Date & Time
df["Date of Purchase"] = pd.to_datetime(df["Date of Purchase"], errors="coerce")
df["Date of Journey"] = pd.to_datetime(df["Date of Journey"], errors="coerce")

time_cols = ["Time of Purchase", "Departure Time", "Arrival Time", "Actual Arrival Time"]
for col in time_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce").dt.time

# Converting Category Columns
categorical_cols = ["Purchase Type", "Payment Method", "Railcard", "Ticket Class", "Ticket Type", "Journey Status", "Refund Request"]
df[categorical_cols] = df[categorical_cols].astype("category")
```

```python
df.groupby('Railcard')['Price'].mean().sort_values(ascending=False)
```

```python
time_cols_to_fix = ['Arrival Time', 'Departure Time', 'Actual Arrival Time']
for col in time_cols_to_fix:
    df[col] = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce').dt.strftime('%H:%M:%S')
df.to_csv('Cleaned_UKRailways_Data_Fixed.csv', index=False, encoding='utf-8')
```
